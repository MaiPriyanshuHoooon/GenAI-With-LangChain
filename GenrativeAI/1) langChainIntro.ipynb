### GROQ Integration

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")

from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("Give me most famous classical pieces of beethoven and mozart")
print(response.content)
```

### Streaming and Batch

```python
for chunk in model.stream("Give me top 5 classicial pieces of beethoven and mozart each"):
    print(chunk.text, end= "|" , flush= True)
```

### Batch

```python
for response in model.batch([
    "What is circle of fifths?",
    "What is the relative minor D major",
    "How does borrowed chord concept work in music theory. Explain by giving example in the key of C"
]):
    print(response)
```
